## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

===========================================================================

- Search Question: Which high-traffic pages are losing performance and need to be updated first?

- Unit of Analysis: A single page (content_hash_id) per client.

- The Decision: Figuring out which pages actually need an edit or refresh, and which ones are doing fine and should be left alone.

- The Action: A content editor opens a ranked list of pages and updates the top ones based on the specific reason code provided (e.g., stale_visible_page).

- The Cost of a Wrong Call:

    False Positive (Flagging a healthy page): Wastes the writing teams expensive hours rewriting content that didn't need to be touched.

    False Negative (Missing a dying page): The page keeps losing ranks, traffic drops to zero, and competitors steal the revenue before anyone notices.

- Why Data/ML Helps: Human editors cant manually check thousands of pages every day. Brittle, hardcoded SQL rules easily miss complex trends
  (like a page with high impressions but dropping scroll depth). ML looks at all these signals at once and ranks the list automatically.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

# uploaded to colab env
df = pd.read_csv('content_refresh_anonymized.csv')

df_cleaned = df[(df['impressions_90d'] > 0) & (df['content_age_days'] >= 90)].drop_duplicates(subset=['content_id'])

total_inventory = len(df_cleaned)
decline_rate = (df_cleaned['trend_direction'] == 'down').mean() * 100
high_demand_decay = len(df_cleaned[(df_cleaned['impressions_90d'] >= 500) & (df_cleaned['trend_direction'] == 'down')])

print(f"--- LANE 2 EVIDENCE METRICS ---")
print(f"Total Unique Filtered Pages: {total_inventory:,}")
print(f"Baseline Decay Rate: {decline_rate:.2f}%")
print(f"High-Demand Assets in Decline: {high_demand_decay:,}")
```
